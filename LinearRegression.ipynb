**Generate 20 Random data points and use linear regression**

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate Random data

```python
np.random.seed(15)

x = np.random.uniform(10, 50, 20)

y = np.random.uniform(100, 1000, 20)

b = 54
```

## Define cost function(MSE)

```python
def compute_cost(w, x, y, b):
    # Number of data points
    n = len(x)
    # Compute Mean Squared Error
    cost = np.sum((y - (w * x + b))**2) / n
    return cost
```

## Compute cost function for different weights

```python
weights = np.linspace(-1000, 1000, 100)
costs = np.zeros_like(weights)

for i, w in enumerate(weights):
    costs[i] = compute_cost(w, x, y,b)
```

```python
best_w_index = np.argmin(costs)
best_w = weights[best_w_index]
```

## Plot-1 the data points

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue', label='Data Points')
```

## Plot-2 Original random data points along with the best-fit line obtained using your Linear Regression model.

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(x, y, color='green', label='Data Points')
x_range = np.linspace(1,50)
y_fit = best_w * x_range + b
plt.plot(x, best_w * x, color='blue', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points and Regression Line')
plt.legend()
```

## Plot-3 cost function vs weight

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 2)
plt.plot(weights, costs, color='green')
plt.xlabel('Weight (w)')
plt.ylabel('Cost (MSE)')
plt.title('Cost Function vs. Weight')

plt.tight_layout()
plt.grid()
plt.show()
```

Then for different values of w we plotted cost function vs weight graph.
