```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate random data for x and y
np.random.seed(0)  # Seed for reproducibility
x = np.random.uniform(10, 50, 20)
y = np.random.uniform(100, 1000, 20)

def compute_cost(x, y, w, b):
    m = len(y)
    cost = (1/(2*m)) * np.sum((w * x + b - y)**2)
    return cost

```

```python
def gradient_descent(x, y, w, b, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        y_pred = w * x + b
        dw = (1/m) * np.sum((y_pred - y) * x)
        db = (1/m) * np.sum(y_pred - y)

        w -= learning_rate * dw
        b -= learning_rate * db

        cost = compute_cost(x, y, w, b)
        cost_history.append(cost)


        if i % 10 == 0:
            plt.scatter(x, y, color='blue')
            plt.plot(x, w*x + b, color='red')
            plt.title(f"Iteration {i} - Cost: {cost:.4f}")
            plt.show()

    return w, b, cost_history
```

```python
# Initialize parameters
w = 0  # Slope
b = 0  # Intercept
learning_rate = 0.0001
iterations = 1000

# Run gradient descent
w_final, b_final, cost_history = gradient_descent(x, y, w, b, learning_rate, iterations)
print(f"Final parameters: w = {w_final}, b = {b_final}")
```

```python

plt.scatter(x, y, color='blue', label="Original Data")

plt.plot(x, w_final*x + b_final, color='red', label="Best Fit Line")
plt.legend()
plt.show()

plt.plot(range(iterations), cost_history)
plt.title("Cost Function J(w) vs Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

```python
import pandas as pd

data = pd.read_csv('/Hyderabad.csv')

data
```

```python
x_housing = data['Area'].values
y_housing = data['Price'].values


x_housing = (x_housing - np.mean(x_housing)) / np.std(x_housing)
y_housing = (y_housing - np.mean(y_housing)) / np.std(y_housing)

def compute_cost(x, y, w, b):
    m = len(y)
    cost = (1/(2*m)) * np.sum((w * x + b - y)**2)
    return cost
```

```python
def gradient_descent(x, y, w, b, learning_rate, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        # Calculate predictions
        y_pred = w * x + b

        # Compute gradients
        dw = (1/m) * np.sum((y_pred - y) * x)
        db = (1/m) * np.sum(y_pred - y)

        # Update parameters
        w -= learning_rate * dw
        b -= learning_rate * db

        # Calculate and store the cost
        cost = compute_cost(x, y, w, b)
        cost_history.append(cost)

        # Plot every few iterations
        if i % 100 == 0:
            plt.scatter(x, y, color='blue', label="Original Data")
            plt.plot(x, w*x + b, color='red', label=f"Fit Line (Iteration {i})")
            plt.legend()
            plt.show()

    return w, b, cost_history
```

```python
w = 0  # Slope (weight)
b = 0  # Intercept (bias)
learning_rate = 0.01
iterations = 1000
```

```python
w_final, b_final, cost_history = gradient_descent(x_housing, y_housing, w, b, learning_rate, iterations)
```

```python
plt.scatter(x_housing, y_housing, color='blue', label="Original Data")
plt.plot(x_housing, w_final * x_housing + b_final, color='red', label="Best Fit Line")
plt.title(f"Best Fit Line: w = {w_final:.4f}, b = {b_final:.4f}")
plt.legend()
plt.show()
```

```python
plt.plot(range(iterations), cost_history)
plt.title("Cost Function J(w) vs Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

#1. Problem Overview:
We are building a Linear Regression model from scratch. This model tries to find a straight line that best represents the relationship between two sets of data (e.g., heights and weights). The goal is to predict the value of one variable (y, like weight) based on another variable (x, like height).

#2. Linear Regression Model:
The line we're trying to find has two key parts:

Slope (w): This tells us how steep the line is, i.e., how much y changes when x changes.
Intercept (b): This tells us where the line crosses the y-axis, i.e., the starting point of the line when x is zero.

#3. Gradient Descent:
We want to adjust the slope (w) and intercept (b) to make the line fit the data as well as possible. To do this, we use a method called gradient descent, which is like a step-by-step process to improve the line.

Think of it like this:

We start with a random line.
In each step, we check how well the line matches the data points.
Based on this, we make small adjustments to the line to make it fit better.
Over time, the line gets closer and closer to being the best fit.

#4. Cost Function:
To know if our line is getting better or worse, we calculate something called cost. The cost tells us how far off our predicted values are from the actual data points. The lower the cost, the better our line fits.

#5. Visualization:
As we adjust the line after each step, we can visualize how it changes. Initially, the line might be far from the data points, but as gradient descent updates the slope and intercept, the line moves closer, eventually becoming a good fit.

By animating this process, you can see the line improving step by step, which helps you understand how the model learns from data.

#In summary:
We start with a random line.
We gradually adjust the line using gradient descent to make it fit the data better.
Each update makes the line more accurate, reducing the difference between predicted and actual values.
The visualization shows how the line evolves to match the data.
